# Generator 


```python
#generator 
'''
Generator is a function that can be a pause and resume 
the program execution 
'''
def my_gen():
    yield 1
    yield 2
    yield 3
    yield 4
for i in my_gen():
    print(i)
```

```python
#example 2
def gen2(n):
    count=1
    while count<=n:
        yield count
        count+=1
for i in gen2(10):
    print(i)
```

```python
#example 3
def gen3():
    yield 1
    yield 2
    yield 3
    yield 4
var=gen3()
print(next(var))
print(next(var))
print(next(var))
print(next(var))

```

```python
#Example 4
list1=[1,2,3,4,5]
iter1=iter(list1)
print(next(iter1))
print(next(iter1))
print(next(iter1))
print(next(iter1))
print(next(iter1))
```

```python
eve = [i for i in range(10) if i%2==0]
for j in eve:
    print(j)
print()
def pr_eve(n):
    for i in range(n):
        if i%2==0:
            yield i
for j in pr_eve(10):
    print(j)
```
