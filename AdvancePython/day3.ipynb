```python
#instance variable
class Student:
    def __init__(self,name,age): #constructor
        self.name=name
        self.age=age
        
        
obj1=Student("suraj",20)
obj2=Student("sachin",21)
print(obj1.name, obj1.age)
```

```python
#instance Mehod
class Student:
    def __init__(self,name,age): #constructor
        self.name=name
        self.age=age
    def display(self):
        print("Name:",self.name)
        print("Age:",self.age,'\n')
obj1=Student("suraj",20)
obj2=Student("sachin",21)
obj1.display()
obj2.display()
```

### class Method 

cls() : is a method not a keyword  is a inbuilt parameter which refers to the class itself. It is used to access class variables and methods within the class.

@ :  is a decorator that is used to define a class method. It indicates that the method is bound to the class and not the instance of the class.

 work with the class variable and use cls
the class method is created using the @classmethod decorator. It takes the class itself as the first argument, conventionally named cls, 
and can access class variables and methods. Class methods are often used for factory methods or to provide alternative constructors for a class.

self keyword cannot use the class method because self refers to the instance of the class, while class methods operate on the class itself.


```python
#class Method 
class student:
    name="suraj"
    @classmethod
    def display(cls):
        print("Name:",cls.name)
obj1=student()
obj1.display()
```

### static method 
static method always declare to class itself u do not need any object to call the static method. It is used to perform a specific task that is related to the class but does not require access to instance variables or methods.

 1. we do not require any class method or instance method to access the static method. It can be called directly using the class name.
 2. does not require the self and cls 
 3. when we accept the static method that time directly access the class name not reuquire the object 

 **@staticmethod** : is a **decorator** that is used to define a static method. It indicates that the method is not bound to the class or instance and can be called without creating an object of the class.



```python
# static method always declare to class itself u do not need any object to call the static method. It is used to perform a specific task that is related to the class but does not require access to instance variables or methods.

class student:
    name="suraj"
    @staticmethod
    def display():
        print("Name:",student.name)
obj1=student()
obj1.display()
```

### Method overriding 
Method Overriding means a child class provides its own implementation of a method that already exists in the parent class.

there is no direct support for method overloading in python same. method name but different paramter

```python
#method overloading
# class Student:
#     def add(a,b,c):
#         return a+b+c
#     def add(a,b):
#         return a+b
# obj1=Student()
# obj1.add(10,20) #causes error because python does not support method overloading. The last defined method will override the previous one.

class Student:
    def add(self,a,b,c=0):
        return a+b+c
    
obj1=Student()
print(obj1.add(10,20))
print(obj1.add(10,20,30))
```

```python
#Method overriding
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def sound(self):      # Overriding parent method
        print("Dog barks")

d = Dog()
d.sound()
```
