```python
#majic method 
''' 
It is a special method which call automatically 
create using def __fun-name__()
ex. __init__()
'''
# str method: represent string
class Student:
    def __init__ (self,name):
        self.name=name
    def __str__(self):
        return f"student ID {self.name}"
std=Student("suraj")
print(std)
```

```python
#len method: used to find length
class Student:
    def __init__(self,names):
        self.names=names
    def __len__(self):
        return len(self.names)
list1=Student(["suraj","lucky","ram"])
print(len(list1))
```

```python
#add method
class Student:
    def __init__(self,num):
        self.num=num
    def __add__(self,var):
        return Student(self.num+var.num)
    def __str__(self):
        return f" {self.num}"
std=Student(10)
obj=Student(20)
print(std+obj)
```
