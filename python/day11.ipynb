```python
#creating set using curly braces 
newset={20,30,40,50,60}
print(newset)

#creating set using set() constructor
newset=set([20,30,40,50,60])
print(newset)

#for creating empty set we have to use set() constructor only 
emptyset=set()
print(emptyset)

```

```python
#sets functions

#for creating empty sets same method just don't give the value... 
newset={10,20,30,40,50,60}
newset.add(70)
print(".add(70):",newset)

#remove() method is used to remove the specified item from the set.
newset.remove(20)
print(".remove(20):",newset)

#discard() method is used to remove the specified item from the set.
#  If the item is not present in the set, it does not raise any error.
newset.discard(20)
print(".discard(20):",newset)

#clears all the values in the set and makes it empty
newset.clear()
print(".clear():",newset,'\n')


#union() method returns a new set with all items from both sets removing duplicates.
newset1={10,20,30,40,50,60}
newset2={40,50,60 ,70,80,90}
result=newset1.union(newset2)
print("newset1.union(newset2):",result)

#intersection() method returns a new set with only the items that are present in both sets.
newset3={10,20,30,40,50,60}
newset4={40,50,60 ,70,80,90}
result=newset3.intersection(newset4)
print("newset3.intersection(newset4):",result)

#difference() method returns a new set with all items from the first set that are not present in the second set.
newset5={10,20,30,40,50,60}
newset6={40,50,60 ,70,80,90}
result=newset5.difference(newset6)
print("newset5.difference(newset6):",result)

#symmetric_difference() method returns a new set with all items from both sets that are not present in both sets.
newset7={10,20,30,40,50,60}
newset8={40,50,60 ,70,80,90}
result=newset7.symmetric_difference(newset8)
print("newset7.symmetric_difference(newset8):",result)

#issubset() method returns True if all items in the set are present in the specified set, otherwise it returns False.
newset9={10,20,30}
newset10={10,20,30,40,50,60} 
result=newset9.issubset(newset10)
print("newset9.issubset(newset10):",result)
```
