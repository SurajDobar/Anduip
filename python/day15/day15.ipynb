```python
# file handling
# open() :-file_name,mode: used to open the file 
# mode:- r(read),w(write),a(append),r+(read and write),a+(append and read),x(create a new file),b(binary mode)
        # w+(write and read),rb(read binary),rw(read and write),ra(read and append),wb(write binary)


# close() :-explicitly close the file 

file = open("day15file_handling.txt","w") # open the file in write mode 
if file:
    print("file opened successfully")
file.close()

```

```python
with open ('day15file.txt','w') as file: # open the file in write mode 
    file.write("Hello World") # write to the file 
    print("w write:file written successfully")

with open('day15file.txt','r') as re: # open the file in read mode 
    content = re.read() # read the file 
    print("r read:",content) # print the content of the file

with open('day15file.txt','r') as file: # open the file in read mode
    for line in file: # read the file line by line 
        print("r read line:",line) # print the content of the file line by line

with open ("day15file.txt",'a') as file: # open the file in append mode 
    file.write("Welcome to python") # append to the file
    print("a append:file appended successfully")
```
