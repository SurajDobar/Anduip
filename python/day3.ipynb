```python
# write a python program to accept the age from the user and check whether the candidate
#  is eligible to vote or not. (age>=18)
age=int(input("Enter your age: "))
if (age>=18):
    print("Eligible to vote")
elif(age>100):
    print("Invalid age")
else:
    print("Not eligible to vote")

    # Enter your age: 20
```

```python
#write a python program to accept the number from the user and check if the number is even or odd
number=int(input("Enter a number: "))
if (number%2==0):
    print("Even")
else:
    print("Odd")

#Enter a number: 2
```

```python
#write a python program to accept alphabet from the user and check if the alphabet is vowel or consonant
alphabet=input("Enter an alphabet: ")   
if alphabet in 'aeiouAEIOU':
    print("Vowel")
else:
    print("Consonant")

#Enter an alphabet: e
```

```python
#write a program to display student result in grade :
# A=above 80 , B=60 to 80 , C= 50 to 60 , D=45 to 50 ,E=35 to 45 , F=below 35
marks=int(input("Enter your marks: "))
if marks>80:
    print("Grade A")
elif marks>=60:
    print("Grade B")
elif marks>=50:
    print("Grade C")
elif marks>=45:
    print("Grade D")
elif marks>=35:
    print("Grade E")
else:
    print("Grade F")
#Enter your marks: 55
```

```python
#write a program for match case (switch)
#1-Monday 2-Tuesday 3-Wednesday 4-Thursday 5-Friday 6-Saturday 7-Sunday
day=int(input("Enter a number between 1 to 7: "))
match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case 6:
        print("Saturday")
    case 7:
        print("Sunday")
    case _:
        print("Invalid input")
    #Enter a number between 1 to 7: 1
```
