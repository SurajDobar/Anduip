```python
#string is a sequence of character enclosed in a single or double quote 
#string is immutable

str="Suraj"
print("Upper case:", str.upper())
print("Lower case:",str.lower(),"\n")

str1="I am learning python programming"
rep=str1.replace("python","java")# .replace is used to replace the substring with another substring
print("Replaced string:",rep,"\n")

str2="I am learning python programming"
print("Split string:",str2.split(" "),"\n")# .split is used to split the string into a list of words

str3=["i","am","learning","python","programming"]
print("Joined string:", "".join(str3),"\n")# .join is used to join the list of words into

#find function is used to find the index occurence of the first character 
str4="I am learning python programming"
print("Index of 'python':", str4.find("python"))
print("Length of the string:", len(str4))# .len is used to find the length of the string
print("Count of 'python':", str4.count("python"),"\n")# .count is used to count the number of occurences of a substring in a string

str5="python"
print("Starts with 'py':", str5.startswith("py"))# .startswith is used to check if the string starts with a substring
print("Ends with 'on':", str5.endswith("on"),"\n")# .endswith is used to check if the string ends
```

```python
str6="        I am learning python programming     "
print("String after stripping whitespace:", str6.strip(),"\n")# .strip is used to remove the whitespace from the beginning and end of the string
print("String after left stripping:", str6.lstrip())# .lstrip is used to remove the whitespace from the beginning of the string
print("String after right stripping:", str6.rstrip(),"\n")# .rstrip is used to remove the whitespace from the end of the string

str7="python class"
print("String contains only alphabetic characters:", str7.isalpha(),'\n')# check all character are alphabetic or not

str8="1234"
print("String contains only digits:", str8.isdigit(),'\n')# check all character are digit or not

str9="python class1234"
print("String contains only alphanumeric characters:", str9.isalnum())# check all character are alphanumeric or not
print("String after removing whitespace:", str9.replace(" ","").isalnum(),'\n')# check all character are alphanumeric or not after removing whitespace

str10="python"
print("Second last character:", str10[-2],'\n')# print the second last character of the string
print("First two characters:", str10[:2])# print the first two characters of the string
print("Characters from index 0 to 2:", str10[0:2])
print("Characters from index 2 to end:", str10[2:])
print("Characters from index 2 to 6:", str10[2:6])
```
