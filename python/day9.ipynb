```python
dict={}#empty dictionary
dict1={'Name':'Zara','course':'Python','id':123}#dictionary with data
print(dict1)#all dictionary data
print(dict1['Name'],'\n')#value retrieved by key

#functions in dictionary
copy_student=dict1.copy() #create a shadow of your dictionary
print('copy function:',copy_student)#copy of dictionary

copy_student.clear()#remove all the key value pair 
print('clear function:',copy_student,'\n')#empty dictionary after clear

student1={'Name':'Zara','course':'Python','id':123}
student_items=student1.items()#return a list of tuple
print('.items() =>',student_items)#print the list of tuple
print('.get() =>',student1.get('Name'))#return the value of associated with the key
print('.keys() =>',student1.keys())#return a list of all the keys
print('.values() =>',student1.values(),'\n')#return a list of all the values

value=student1.pop('course')#remove the key value pair and return the value
print('.pop() =>',value)#print the value of the removed key value pair
print(student1,'\n')#print the dictionary after removing the key value pair

dict1={"emp_id":101,"salary":10000,"dept":"IT"}
item=dict1.popitem()#remove the last inserted key value pair and return it as a tuple
print('.popitem() =>',item)#print the tuple of the removed key value pair
```

```python
#traversing the dictionary
dict2={'Name':'Zara','course':'Python','id':123}

print('keys of the dictionary:')
for key in dict2:
    print(key)#returns the key of the dictionary

print()

print("values for dictionary:")
for values in dict2.values():
    print(values)#returns the values of the dictionary

print()

print("key and values for dictionary:")
for key,values in dict2.items():
    print(key,values)#returns the key and values of the dictionary
print()

dict3={"Name":"suraj"}
dict3["Name"]="suraj dobar" #update the value of the key
print("updated dictionary:",dict3,'\n')#print the updated dictionary

dict4={"Name":"suraj","course":"Python","Name":"suraj dobar"} #duplicate key will be overwritten by the last value
print("updated dictionary with duplicate key:",dict4)#print the dictionary with updated value of the duplicate key

dict5={"suraj":99,"vikas":-99,"sachin":100}
print("dictionary with multiple keys:",dict5,'\n')#print the dictionary with multiple keys

#nested dictionary
students={
    "101":{"Name":"suraj","course":"Python"},
    "102":{"Name":"vikas","course":"Java"},
}
print(students["101"]["Name"],"\n")#accessing the value of the nested dictionary

dict6={"suraj":99,"vikas":-99,"sachin":100}
if "Aarish" in dict6:
    print("Aarish is present in the dictionary")
else :
    print("key not exist")

```
